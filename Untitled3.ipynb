```python

def tryint(s):
    try:
        return int(s)
    except:
        return s

def alphanum_key(s):
    return [tryint(c) for c in re.split('([0-9]+)', s)]

def sort_nicely(l):
    l.sort(key=alphanum_key)
    return l


directory = "depth"
fileNameKeys = ['File','Names','Contain','These','Strings']


dirList = sort_nicely([file for file in os.listdir(directory) if all(substring in file for substring in fileNameKeys)])
images = [cv2.imread(directory + '/' + imageFile) for imageFile in dirList if True]
print(dirList)


#################################
stitcher =cv2.Stitcher_create()
#################################


stitchOut = stitcher.stitch(images)
print("Stitcher exit status = " + str(stitchOut[0]))
print("Hit any Key to continue and save image")
plt.figure(figsize=[25,25])
plt.imshow(stitchOut[0]);plt.title("Matches");plt.axis('off');
#cv2.imwrite(directory + "/" + "STITCH.png", stitchOut[1])
```

```python
a=cv2.imread('dd/0000.jpg',1)
b=cv2.imread('dd/0001.jpg',1)
c=cv2.imread('dd/0002.jpg',1)
d=cv2.imread('dd/0003.jpg',1)
```

```python
hsv = cv2.cvtColor(a, cv2.COLOR_BGR2HSV)
lower_red = np.array([-9,245,215])
upper_red = np.array([180,255,255])
mask = cv2.inRange(hsv, lower_red, upper_red)
fgmask = cv2.erode(mask,None,iterations = 5)
contours, heirarchy = cv2.findContours(fgmask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
for cnt in contours:
    if cv2.contourArea(cnt) >1000:
        x,y,w,h = cv2.boundingRect(cnt)
        cv2.rectangle(fgmask,(x ,y),(x+w,y+h),(100,150,200),3)
        cv2.putText(fgmask,'Object Detected',(x,y+20), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (120,255,80), 1, cv2.LINE_AA)
#cv2.imwrite('panora.jpg', fgmask)
plt.figure(figsize=[25,25])
plt.imshow(fgmask[...,::-1]);plt.title("Matches");plt.axis('off');
```
